```python
import pandas as pd
```

```python
dadosTT = pd.read_csv('titanic.csv')
```

```python
dados_filtrados = dadosTT.loc[
    (dadosTT['Sex'] == 'male') &
    (dadosTT['Pclass'] == 1 ) &
    (dadosTT['Survived'] == 1 ),

    ['Name','Age','Sex','Pclass','Survived']
    ] .sort_values(by='Name')
dados_filtrados
```

```python
dados_filtrados.to_csv('titanic_resultado.csv', index = False)
```

```python
dados_filtrados.to_excel('titanic_resultado.xlsx', index = False)
```
